```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
inputs0 = np.array([0, 0]) + np.random.randn(10, 2)
inputs1 = np.array([0, 3]) + np.random.randn(10, 2)
inputs2 = np.array([2, 2]) + np.random.randn(10, 2)
inputs = np.concatenate([inputs0, inputs1, inputs2], axis=0)
labels = np.array([0]*10 + [1]*10 + [2]*10)
ind = np.random.permutation(30)
inputs = inputs[ind, :]
labels = labels[ind]
```

```python
plt.plot(inputs[labels == 0, 0], inputs[labels == 0, 1], "b.")
plt.plot(inputs[labels == 1, 0], inputs[labels == 1, 1], "r.")
plt.plot(inputs[labels == 2, 0], inputs[labels == 2, 1], "g.")
```

```python
inputs = np.concatenate((np.ones((30, 1)), inputs), axis=1)
```

```python
def perceptron_epoch(inputs, labels, w, eta):
  #print(w)
  mistakes = 0

  for x,y in zip(inputs, labels):
    y_hat = np.argmax(w.dot(x))
    if(y_hat != y):
      w[y] += eta*x
      w[y_hat] -= eta*x
      mistakes += 1

  print('Mistakes: ', mistakes)
  print('Accuracy: ', round((1-(mistakes/30))*100,2), '%')
  print('----')
  return w
    
```

```python
w = np.zeros((3, 2+1))

for i in range(100): 
  #print(w)
  print('Epoch:', i)
  w = perceptron_epoch(inputs, labels, w, 1)
```
